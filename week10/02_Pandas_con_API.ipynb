````python
import datetime as dt
import pandas as pd
import requests
````

````python
# Declaración de variables
ayer = dt.date.today() - dt.timedelta(days=1)
api_url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
payload = {
    'format' : 'geojson'
    ,'starttime': ayer - dt.timedelta(days=30)
    ,'endtime' : ayer
}
````

````python
# Creamos la petición
response = requests.get(api_url, params=payload )
response.status_code
````

**status_code** - respuestas del servidor
- 200 OK
- 3xx redirección del recurso
- 4xx recurso no encontrado
- 5xx error servidor

````python
# El servidor devuelve un encabezado (headers) y un cuerpo (content)
response.headers
````
