min() 	The minimum of the values 	Numerical
idxmin() 	The index where the minimum values occurs 	Numerical
max() 	The maximum of the values 	Numerical
idxmax() 	The index where the maximum value occurs 	Numerical
abs() 	The absolute values of the data 	Numerical
std() 	The standard deviation 	Numerical
var() 	The variance 	Numerical
cov() 	The covariance between two Series, or a covariance matrix for all column combinations in a DataFrame 	Numerical
corr() 	The correlation between two Series, or a correlation matrix for all column combinations in a DataFrame 	Numerical
quantile() 	Calculates a specific quantile 	Numerical
cumsum() 	The cumulative sum 	Numerical or Boolean
cummin() 	The cumulative minimum 	Numerical
cummax() 	The cumulative maximum 	Numerical

````python
# Operamos con expresiones regulares
df['place'].str.extract(r', (.*$)')[0].sort_values().unique() # recoge los valores después de la coma y espacio
````

````python
df['place']
````

````python
# Operamos con transformación de unos valores
df['parsed_place'] = df['place'].str.replace(
    r'.* of ', '', regex=True # elimina cualquier término que tenga of
).str.replace(
    'the ', '' # eliminamos "the "
).str.replace(
    r'CA$', 'California', regex=True # sustituye CA por California
).str.replace(
    r'NV$', 'Nevada', regex=True # sustituye NV por Nevada
).str.replace(
    r'MX$', 'Mexico', regex=True # sustituye MX por Mexico
).str.replace(
    r' region$', '', regex=True # elimina el término region
).str.replace(
    'northen ', '' # eliminamos 'northen ' 
).str.replace(
    'Fiji Islands', 'Fiji' # renombramos Fiji Island por Fiji
).str.replace(
    r'^.*, ', '', regex=True # elimina cualquier elemento raro desde el principio
).str.replace(
    r'Fiji Islan[a-zA-Z0-9]{1,}', 'Fiji', regex=True
).str.strip() # elimina espacios
````

````python
df['parsed_place'].sample(20, random_state=6)
````

````python
df[['place','parsed_place']].sample(20,random_state=20)
````

````python
# Podemos hacer uso de assign() para crear columnas
df.assign(
    in_ca=df['parsed_place'].str.endswith('California'),
    in_alaska=df['parsed_place'].str.endswith('Alaska')
).sample(20, random_state=20)
````
